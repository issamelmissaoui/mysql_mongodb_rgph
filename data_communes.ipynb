## 1.4 Langues Locales

get fields of languages table

```python
url = f"http://rgphentableaux.hcp.ma/Default1/getDATA/?type=Commune&CGEO={communes.loc[0,'Code_Commune']}&them=5"
result = requests.get(url, headers={'Referer': 'http://rgphentableaux.hcp.ma/Default1/'})
len(result.json())
```

```python
def replaceMultiple(mainString, toBeReplaces, newString):
    # Iterate over the strings to be replaced
    for elem in toBeReplaces :
        # Check if string is in the main string
        if elem in mainString :
            # Replace the string
            mainString = mainString.replace(elem, newString)
    
    return  mainString
```

```python
lang_fields = [None]*45
for i in range(0,45) :
    if len(result.json()[i]['INDICATEUR']) > 63 :
        lang_fields[i] = replaceMultiple(result.json()[i]['INDICATEUR'][:60], ["'", " ", '-', '(', ')'], "_")
    else :
        lang_fields[i] = replaceMultiple(result.json()[i]['INDICATEUR'], ["'", " ", '-', '(', ')'], "_")
```

```python
lang_fields
```

```python
data_lang = pd.DataFrame(np.random.randint(0,100,size=(1687, 45)), columns=lang_fields)
data_lang['id_comm'] = data_lang[lang_fields[0]].astype(str)
data_lang = data_lang[ ['id_comm'] + [ col for col in data_lang.columns if col != 'id_comm' ] ]
data_lang.head()
```

```python
for i in range(0,1687) : 
    url = f"http://rgphentableaux.hcp.ma/Default1/getDATA/?type=Commune&CGEO={communes.loc[i,'Code_Commune']}&them=5"
    result = requests.get(url, headers={'Referer': 'http://rgphentableaux.hcp.ma/Default1/'})
    data_lang.loc[i,'id_comm'] = communes.loc[i,'Code_Commune']
    for j in range(0,45) :
        if (result.json()[j]['DATA2014'] == '-') or (result.json()[j]['DATA2014'] == '/') :
            data_lang.loc[i,lang_fields[j]] = 0
        else :
            data_lang.loc[i,lang_fields[j]] = float(result.json()[j]['DATA2014'])
```

```python
data_lang.to_excel('data_lang.xlsx')
```
